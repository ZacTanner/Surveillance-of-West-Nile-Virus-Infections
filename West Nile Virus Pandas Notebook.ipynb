```python
#Import the EDA packages
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#Load the Data Frame
df = pd.read_csv('West Nile.csv')
```

```python
df
```

```python
#Checking for null entries
df.isna().sum()
```

```python
#All the NaN values in the New Places with human cases correspond to 'no' indication from the data. Data is not actually missing, so will convert all values to 'No'.
df[['New places with human cases']] = df[['New places with human cases']].fillna('No')
```

```python
df['New places with human cases'].value_counts()
```

```python
df['TimeCode first human infection reported current season (yyyy-mm-dd)'].value_counts()
```

##### Nothing seems off with the value counts above.

```python
#Renaming the fields to simplify analysis.
df = df.rename(columns={
    'Place of infection (NUTS3/Gaul1)': 'Location Code',
    'Place of infection name': 'Place of Infection',
    'TimeCode first human infection reported current season (yyyy-mm-dd)': 'Reported Infection Date',
    'New places with human cases': 'Newly Affected Region',
    'Data as of': 'Year and Week Number'
})
```
